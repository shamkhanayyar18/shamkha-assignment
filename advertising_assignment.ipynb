```python
import pandas as pd
```

```python
df = pd.read_csv("advertising (1).csv")
```

```python
df.drop("Newspaper", axis=1, inplace= True)
```

```python
df.head()
```

```python
df.shape
```

```python
x= df [['TV','Radio']]
y= df [["Sales"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=18)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
sqrt(mean_squared_error(y_train, y_pred_train))
```

```python
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
sqrt(mean_squared_error(y_test, y_pred_test))
```

```python
r2_score(y_test, y_pred_test)
```

```python
def sales_predict(tv, radio):
    predicted_sale = model.predict([[tv, radio]])
    return predicted_sale[0][0]
```

```python
tv = float(input('Enter budget for tv: '))
radio = float(input('Enter budget for radio: '))

print(f"Predicted Sale: {round(sales_predict(tv, radio))}")
```
