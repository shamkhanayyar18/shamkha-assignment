```python
import pandas as pd
```

```python
df = pd.read_csv("Student_Performance (1).csv")
df.head()
```

```python
x= df[['Previous Scores']]
y = df[['Performance Index']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
Performance_Index = LinearRegression()
```

```python
Performance_Index.fit(X_train,y_train)
```

```python
Performance_Index.coef_
```

```python
Performance_Index.intercept_
```

```python
# Prediction on Training Data
```

```python
y_pred_train = Performance_Index.predict(X_train)
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(kind='scatter', x='Previous Scores', y='Performance Index',
       title='studiet hour pricdiction')
```

```python
# pridicted line
```

```python
plt.scatter(x=X_train, y=y_train, color='red')
plt.scatter(x=X_train, y=y_pred_train, color='yellow')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_train, y_pred_train)
```

```python
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = Performance_Index.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred_test)
```

```python
r2_score(y_test, y_pred_test)
```
